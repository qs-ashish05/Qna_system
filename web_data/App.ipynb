```python
from bs4 import BeautifulSoup
import requests
from urllib.request import urlopen
import pandas as pd
```

```python
def fetch_data(url):
    Products_name = []
    number_of_rating_stars = []
    number_of_rating = []
    number_of_reviews = []
    current_prices = []
    Processors = []
    Ram = []
    Os = []
    ssd = []
    Display = []
    Warranty = []
    image_url = []


    url = url
    html = urlopen(url)
    html_doc = html.read()

    soup = BeautifulSoup(html_doc)

    Products = soup.find_all('div',class_ = '_2kHMtA')

    for i in Products:

        name = i.find('div',class_ = '_4rR01T').text.split('-')[0].strip() 
        Products_name.append(name)

    # Finding rating stars 

        temp = i.find('div',class_ = '_3LWZlK')
        try:
            star = temp.text
        except:
            star = 0
        number_of_rating_stars.append(star)


        # finding number of revies and ratings
        Rat_Rev = i.find('span',class_ = '_2_R_DZ')
        #print(Rat_Rev.text)
        try:
            list = Rat_Rev.text.split('&')
            Rating = list[0].strip()
            Review = list[1].strip()
            rating_count = Rating.split()[0]
            review_count = Review.split()[0]
        except:
            rating_count = 0
            review_count = 0
        # print(rating_count,review_count)
        number_of_rating.append(rating_count)
        number_of_reviews.append(review_count)


        # extracting price 
        current_price = i.find('div',class_ = '_30jeq3 _1_WHN1')
        #print(current_price.text)
        list = current_price.text.split('â‚¹')
        #print(list)
        current_price = list[1]
        list1 = current_price.split(',')
        #print(list1)
        current_price = list1[0] + list1[1]
        #print(current_price)
        current_prices.append(current_price)


        info_temp = i.findAll('li',class_ = 'rgWa7D')  # rgWa7D
        try : 
            proc_info = info_temp[0].text
            ram_info = info_temp[1].text
            os_info = info_temp[2].text
            ssd_info = info_temp[3].text
            display_info = info_temp[4].text
            warr_info = info_temp[5].text
        except : 
            proc_info = None
            ram_info = None
            os_info = None
            ssd_info = None
            display_info = None
            warr_info = None
        
        Processors.append(proc_info)
        Ram.append(ram_info)
        Os.append(os_info)
        ssd.append(ssd_info)
        Display.append(display_info)
        Warranty.append(warr_info)

        # Finding image url
        imgage = i.find('img')           
        imgurl = imgage.get('src')     
        image_url.append(imgurl)

    dict = {
    'Product Name': Products_name,
    'Number_of_rating_stars':number_of_rating_stars,
    'number_of_rating':number_of_rating,
    'number_of_reviews':number_of_reviews,
    'current_prices':current_prices,
    'Processors':Processors,
    'Ram':Ram,
    'Os':Os,
    'SSD':ssd,
    'Display':Display,
    'Warranty':Warranty,
    'image_url':image_url
    }

    df = pd.DataFrame(dict)
    return df
```

## Fetching data of laptops

```python
url = f"https://www.flipkart.com/search?q=mobile%20phones&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
data = fetch_data(url)
```
